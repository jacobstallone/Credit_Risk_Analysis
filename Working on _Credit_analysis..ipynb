## Introduction 

Running the bank vs being the bank.

This project is aimed at assessing the credit worthiness of a borrower to make an informed
decision on lending,credit risk and predicting potential default.
Every bank faces two types of risk when there is a new client that comes through the door.
Going through the client's profile, credit history and portfolio is an intense process before the
bank can make a decision on lending.If the applicant is a good credit risk, i.e. is likely to repay
the loan, then not approving the loan to the person results in losing business,client relationship
and customer goodwill. On the other hand, If the applicant is a bad credit risk, i.e. is not likely to
repay the loan, then approving the loan could potentially result in payment default and financial
loss to the bank.

Data exploration and analysis can help knowledge revealing patterns in data with respect to-client default
behavior, repayment capacity, regularity of income, socio-economic factors, assets and
liabilities, loan tenure, risk threshold,etc. It is important to understand the rationale behind
these components to protect bank&#39;s interests and enhance client service.Ultimate goal being
aligning both objectives for minimization of risk and maximization of profit.

examples to use:
Individual with multiple income sources, high value assets, multiple number of defaults -low risk
individual with low income, no assets,regular high payment-potential high risk

The German Credit Data contains data on 20 variables and the classification whether an
applicant is considered a Good or a Bad credit risk for 1000 loan applicants. A predictive model developed on this data
is expected to provide a bank manager guidance for making a decision whether to approve a
loan to a prospective applicant based on his/her profiles.

```python
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_val_score
scorer = make_scorer(mean_squared_error, greater_is_better = False)

def rmse_cv_train(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y_train, scoring = scorer, cv = 10))
    return(rmse)

def rmse_cv_test(model):
    rmse= np.sqrt(-cross_val_score(model, X_test, y_test, scoring = scorer, cv = 10))
    return(rmse)
```

Lasso Regression Feature Selection:

```python
#lasso
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV
lasso = LassoCV(alphas = [0.0001, 0.0003, 0.0006, 0.001, 0.003, 0.006, 0.01, 0.03, 0.06, 0.1, 
                          0.3, 0.6, 1], 
                max_iter = 50000, cv = 10)
lasso.fit(X_train, y_train)
alpha = lasso.alpha_
print("Best alpha :", alpha)

print("Try again for more precision with alphas centered around " + str(alpha))
lasso = LassoCV(alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, 
                          alpha * .85, alpha * .9, alpha * .95, alpha, alpha * 1.05, 
                          alpha * 1.1, alpha * 1.15, alpha * 1.25, alpha * 1.3, alpha * 1.35, 
                          alpha * 1.4], 
                max_iter = 50000, cv = 10)
lasso.fit(X_train, y_train)
alpha = lasso.alpha_
print("Best alpha :", alpha)

print("Lasso RMSE on Training set :", rmse_cv_train(lasso).mean())
print("Lasso RMSE on Test set :", rmse_cv_test(lasso).mean())
y_train_las = lasso.predict(X_train)
y_test_las = lasso.predict(X_test)
```

### stochastic gradient descent

####  Classification with stochastic gradient descent

The `SGDClassifier` is very similar to the `SGDRegressor`. The main difference is that the loss functions are changed to regression loss functions.

**Arguments**
- `loss`: `['log', ...]`
    - The `'log'` loss corresponds to solving a logistic regression classifier.
- `penalty`: `['none','l1','l2','elasticnet']`
    - This defines the penalty on the regression that you would like to solve. The l1 and l2 are the Lasso and Ridge, while the elasticnet is the combination of them both.
- `alpha`
    - The regularization strength to be used with a chosen penalty. Same as in Lasso and Ridge.
- `l1_ratio`
    - The mix of the Lasso and Ridge penalties when elasticnet is chosen as the penalty.
- `n_iter`
    - The number of training "epochs" over the data. This is the number of passes that the gradient descent algorithm will make over the data to iteratively fit the weights (defaults to 5).

Like `SGDRegressor`, `SGDClassifier` is most often used in tandem with grid searching to find the optimal parameters for certain models

```python
from sklearn.linear_model import SGDRegressor, SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.grid_search import GridSearchCV
```

```python
sgd_cls_params = {
    'loss':['log'],
    'penalty':['l1','l2'],
    'alpha':np.logspace(-5,2,50)
}

sgd_cls = SGDClassifier()
sgd_cls_gs = GridSearchCV(sgd_cls, sgd_cls_params, cv=5, verbose=1)
```

In a ROC curve, the true positive rate (Sensitivity) is plotted as a function of the false positive rate (100-Specificity) for different cut-off points of a parameter.This is not very realistic, but it does mean that a larger area under the curve (AUC) is usually better.



```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(y_test, yhat_pp[:,1])
roc_auc = auc(fpr, tpr)
fpr_lasso, tpr_lasso, _ = roc_curve(y_test, yhat_lasso_pp[:,1])
roc_auc_lasso = auc(fpr_lasso, tpr_lasso)
fpr_ridge, tpr_ridge, _ = roc_curve(y_test, yhat_ridge_pp[:,1])
roc_auc_ridge = auc(fpr_ridge, tpr_ridge)
```

```python
plt.figure(figsize=[8,8])

plt.plot(fpr, tpr, label='Logistic (area = %0.2f)' % roc_auc, linewidth=4)
plt.plot(fpr_ridge, tpr_ridge, label='Ridge (area = %0.2f)' % roc_auc_ridge, 
         linewidth=4, color='darkred')
plt.plot(fpr_lasso, tpr_lasso, label='Lasso (area = %0.2f)' % roc_auc_lasso, 
         linewidth=4, color='darkgreen')

plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([-0.05, 1.0])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('Receiver operating characteristic: is male', fontsize=18)
plt.legend(loc="lower right")
plt.show()
```

The area under the ROC curve (AUC) is a measure of how well a parameter can distinguish between two diagnostic groups (Risk Vs No Risk). Based on this plot, the area under the curve for our best model was found to be ~0.83. This shows that a randomly selected observation from the training data set with a '1' (not likely to risk) label has a score larger than that for a randomly chosen observation from the group with the '0' (Risk) label 83% of the time.
